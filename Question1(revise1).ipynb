```python
import numpy as np
from tqdm import trange
```

```python
#given parameters
c = 0.055 #Coupon rate
T = 5  #Maturity
Call_Price = 103
stepsize = 0.5  #Semi-Annual
initial_rate = 0.06
u = 1.25
d = 1/u
q = 0.5
p = 1-q
N = int(T/stepsize)
price = 250*(1+c) #Price at maturity time
```

### **Compute Bond Price**

```python
short_rate = np.zeros((N+1,N+1)) 
print(short_rate)
```

```python
short_rate[0][0] = initial_rate
print(short_rate)
```

```python
for i in trange(len(short_rate)):
    for j in trange(len(short_rate)):
      if i > j :
          pass
      else:
          short_rate[i][j] =(u**i)*(d**(j-i))*initial_rate
print(short_rate)
del(i,j)
```

```python
Price = np.zeros((N+1,N+1))
print(Price)
```

```python
for i in trange(len(Price)-1,-1,-1):
    for j in trange(len(Price)-1,-1,-1):
      if i > j :
          pass
      else:
          if j == N:
              Price[i][j] = 250
          else:
              Price[i][j] = (q*Price[i+1][j+1] + p*Price[i][j+1])/(1+short_rate[i][j])
print(Price)
Bond_Price = Price[0][0]
print('Bond Price = {}'.format(round(Bond_Price,2)))
```

### **Compute Callable Bond Price**

```python
short_rate = np.zeros((N+1,N+1))
short_rate[0][0] = initial_rate
for i in trange(len(short_rate)):
    for j in trange(len(short_rate)):
      if i > j :
          pass
      else:
          short_rate[i][j] =(u**i)*(d**(j-i))*initial_rate
#print(short_rate)
del(i,j)

Price = np.zeros((N+1,N+1))
for i in trange(len(Price)-1,-1,-1):
    for j in trange(len(Price)-1,-1,-1):
      if i > j :
          pass
      else:
          if j == N:
              Price[i][j] = price
          else:
              Price[i][j] = 100*c + (q*Price[i+1][j+1] + p*Price[i][j+1])/(1+short_rate[i][j])
print(Price)
Bond_Price = Price[0][0]
Callable_Bond_Price = round(Bond_Price - Call_Price,2)
print('Callable Bond Price = {}'.format(Callable_Bond_Price))
```

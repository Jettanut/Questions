```python
from __future__ import division
import sys
import math 
from scipy.optimize import fsolve
import pandas as pd
import numpy as np
from tqdm import trange
```

```python
#given parameters
c = 0.055 #Coupon rate
T = 5  #Maturity
Call_Price = 103
stepsize = 0.5  #Semi-Annual
initial_rate = 0.06
u = 1.25
d = 1/u
q = 0.5
p = 1-q
N = int(T/stepsize)
price = 250*(1+c) #Price at maturity time
```

### **Compute Bond Price**

```python
short_rate = np.zeros((N+1,N+1)) 
print(short_rate)
```

```python
short_rate[0][0] = initial_rate
print(short_rate)
```

```python
for i in trange(len(short_rate)):
    for j in trange(len(short_rate)):
      if i > j :
          pass
      else:
          short_rate[i][j] =(u**i)*(d**(j-i))*initial_rate
print(short_rate)
del(i,j)
```

```python
Price = np.zeros((N+1,N+1))
print(Price)
```

```python
for i in trange(len(Price)-1,-1,-1):
    for j in trange(len(Price)-1,-1,-1):
      if i > j :
          pass
      else:
          if j == N:
              Price[i][j] = 250
          else:
              Price[i][j] = (q*Price[i+1][j+1] + p*Price[i][j+1])/(1+short_rate[i][j])
print(Price)
Bond_Price = Price[0][0]
print('Bond Price = {}'.format(round(Bond_Price,2)))
```

```python
def bdtOne(guess):
    r_u = guess * math.exp(2 * df.Vol.iloc[1])
    r_d = guess 
    N1 = (100)/(1+r_u)
    N2 = (100)/(1+r_d)
    return (0.5*((N1/(1+r[0])) + (N2/(1+r[0])))-a2)
    
g = fsolve(bdtOne,m)[0]
r_u = g * math.exp(2 * df.Vol.iloc[1])
r_d = g
```

```python
def bdtTwo2(x):
    r_u2 = x[0] * math.exp(4 * x[1])
    r_ud = x[0] * math.exp(2 * x[1])
    r_d2 = x[0]
    N1 = (100)/(1+r_u2)
    N2 = (100)/(1+r_ud)
    N3 = (100)/(1+r_d2)
    ans1 = (0.5*(N1 + N2))/(1+r_u)
    ans2 = (0.5*(N2 + N3))/(1+r_d)
    y_u = math.sqrt(100/ans1) - 1
    y_d = math.sqrt(100/ans2) - 1
    out = [((0.5*(ans1 + ans2))/(1+r[0])-a3)]
    out.append((0.5 * math.log(y_u/y_d) - df.Vol.iloc[2]))
    return out

k = fsolve(bdtTwo2,[0.1,0.1])
r_u2 = k[0] * math.exp(4 * k[1])
r_ud = k[0] * math.exp(2 * k[1])
r_d2 = k[0]
```
